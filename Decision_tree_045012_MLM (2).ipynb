```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import time
import psutil

# Step 1: Load data
file_path = "100000 Sales Records.xlsx"
data = pd.read_excel(file_path)

# Step 2: Data Preprocessing
# Convert categorical variables to one-hot encoding
data_encoded = pd.get_dummies(data, columns=['Region', 'Country', 'Item Type', 'Sales Channel'])
data_encoded['Order Priority'] = pd.Categorical(data_encoded['Order Priority'], ordered=True, categories=['C', 'H', 'L', 'M'])
data_encoded['Order Priority'] = data_encoded['Order Priority'].cat.codes

# Drop unnecessary columns
data_preprocessed = data_encoded.drop(columns=['Order ID', 'Order Date', 'Ship Date'])

# Step 3: Split Data
X = data_preprocessed.drop(columns=["Order Priority"])
y = data_preprocessed["Order Priority"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train Decision Tree Classifier
clf_decision_tree = DecisionTreeClassifier()
start_time = time.time()
clf_decision_tree.fit(X_train, y_train)
end_time = time.time()
y_pred_decision_tree = clf_decision_tree.predict(X_test)

# Step 5: Evaluate Decision Tree Classifier
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)
report_decision_tree = classification_report(y_test, y_pred_decision_tree)
matrix_decision_tree = confusion_matrix(y_test, y_pred_decision_tree)
memory_decision_tree = psutil.Process().memory_info().rss
execution_time_decision_tree = end_time - start_time

# Display results
print("Decision Tree Classifier Results:")
print("Accuracy:", accuracy_decision_tree)
print("Classification Report:")
print(report_decision_tree)
print("Confusion Matrix:")
print(matrix_decision_tree)
print("Execution Time:", execution_time_decision_tree)
print("Memory Usage:", memory_decision_tree)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import seaborn as sns  # Adding this import statement

# Load the data
data = pd.read_excel("100000 Sales Records.xlsx")

# Data Pre-processing
# Convert 'Order Priority' to numerical values
data['Order Priority'] = pd.Categorical(data['Order Priority'], ordered=True, categories=['C', 'H', 'L', 'M'])
data['Order Priority'] = data['Order Priority'].cat.codes

# Define features and target variable
X = data[['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']]
y = data['Order Priority']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train Decision Tree model
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)

# Train Logistic Regression model
log_reg_classifier = LogisticRegression(random_state=42)
log_reg_classifier.fit(X_train, y_train)

# Train SVM model
svm_classifier = SVC(kernel='linear', random_state=42)
svm_classifier.fit(X_train, y_train)

# Train KNN model
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train, y_train)

# Evaluate models
models = {
    'Decision Tree': dt_classifier,
    'Logistic Regression': log_reg_classifier,
    'SVM': svm_classifier,
    'KNN': knn_classifier
}

for name, model in models.items():
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    matrix = confusion_matrix(y_test, y_pred)

    print(f"Model: {name}")
    print("Accuracy:", accuracy)
    print("Classification Report:")
    print(report)
    print("Confusion Matrix:")
    print(matrix)
    print()



# Define a function to plot confusion matrix
def plot_confusion_matrix(model_name, y_test, y_pred):
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

# Evaluate models and plot confusion matrices for KNN, SVM, and Logistic Regression
models_to_evaluate = {
    'KNN': knn_classifier,
    'SVM': svm_classifier,
    'Logistic Regression': log_reg_classifier
}

for name, model in models_to_evaluate.items():
    y_pred = model.predict(X_test)
    plot_confusion_matrix(name, y_test, y_pred)
```
